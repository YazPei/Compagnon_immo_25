```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importer les fichiers
df_loc = pd.read_csv('/Users/maximehenon/Documents/GitHub/MAR25_BDS_Compagnon_Immo/ech_annonces_locations_68.csv', sep=';' ,  on_bad_lines='skip')
df_vent = pd.read_csv('/Users/maximehenon/Documents/GitHub/MAR25_BDS_Compagnon_Immo/ech_annonces_ventes_68.csv', sep =';',on_bad_lines='skip')

# Liste des colonnes:
display(df_loc.columns)
display(df_vent.columns)
```

```python
# Liste des colonnes num√©riques des datasets

NUMERIC_COLUMNS_loc = ['etage','surface', 'surface_terrain','prix_bien',
       'prix_maison', 'prix_terrain', 'mensualiteFinance', 'dpeC', 'places_parking', 'annee_construction', 'surface_balcon', 'nb_logements_copro','charges_copro']
NUMERIC_COLUMNS_vent = ['etage','surface', 'surface_terrain','prix_bien','prix_maison', 'prix_terrain', 'mensualiteFinance', 'dpeC', 'places_parking', 'annee_construction', 'surface_balcon', 'nb_logements_copro','charges_copro','loyer_m2_median_n6', 'taux_rendement_n6',
       'loyer_m2_median_n7', 'taux_rendement_n7','prix_m2_vente']

```

```python
# Afficher les statistiques descriptives colonne par colonne pour le dataset des locations
print("Statistiques descriptives pour le dataset des locations (colonne par colonne):")
for colonne in NUMERIC_COLUMNS_loc:
    print(f"\nColonne : {colonne}")
    print(df_loc[colonne].describe())
    print(f"Variance : {df_loc[colonne].var()}")
    print(f"Somme : {df_loc[colonne].sum()}")
    pd.options.display.float_format = '{:.2f}'.format

# Afficher les statistiques descriptives colonne par colonne pour le dataset des ventes
print("\nStatistiques descriptives pour le dataset des ventes (colonne par colonne):")
for colonne in NUMERIC_COLUMNS_vent:
    print(f"\nColonne : {colonne}")
    print(df_vent[colonne].describe())
    print(f"Variance : {df_vent[colonne].var()}")
    print(f"Somme : {df_vent[colonne].sum()}")
    pd.options.display.float_format = '{:.2f}'.format
    
```

```python
# fonction pour tracer les boxplots :
def tracer_boxplots(df, colonnes):
    for colonne in colonnes:
        plt.figure(figsize=(8, 6))
        plt.boxplot(df[colonne])
        plt.title(f'Boxplot de {colonne}')
        plt.ylabel(colonne)
        plt.show()
```

```python
# tracer les boxplots pour les colonnes num√©riques
tracer_boxplots(df_loc, NUMERIC_COLUMNS_loc)
tracer_boxplots(df_vent, NUMERIC_COLUMNS_vent)
```
